```python
plt.figure(figsize=(40,40))
plotnumber=1
for i in data1:
    plt.subplot(4,4,plotnumber)
    sns.boxplot(x=data[i])
    plotnumber=plotnumber+1
plt.tight_layout()
```

```python
def find_boundaries(data,variables):
    Q1=data[variables].quantile(0.25)
    Q3=data[variables].quantile(0.75)
    IQR=Q3-Q1
    lowerboundary=Q1-1.5*IQR
    upperboundary=Q3+1.5*IQR
    return  lowerboundary,upperboundary
```

```python
lower_MinTemp,upper_MinTemp=find_boundaries(data,'MinTemp')
print('upper bondary for MinTemp',upper_MinTemp)
print('lower boundary for MinTemp',lower_MinTemp)
```
